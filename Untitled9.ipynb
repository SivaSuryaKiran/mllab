```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("diabetes.csv")
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
data.iloc[:,0:8] = mms.fit_transform(data.iloc[:,0:8])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:,0:8]
```

```python
y = data.iloc[:,8:9]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=124)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(X_train,y_train)
```

```python
pred = DTC.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,r2_score
```

```python
confusion_matrix(pred,y_test)
```

```python
precision_score(pred,y_test)
```

```python
r2_score(pred,y_test)
```

```python
recall_score(pred,y_test)
```
